```python
import pandas as pd
```

```python
df = pd.read_csv('lc_ds.csv')
```

```python
dff = df[df['difficulty'] == 'Hard']
dff = dff[['id', 'description']]
dff.to_csv('hard_problems.csv', index=False)
```

```python
import json

dff_ids = dff['id'].tolist()
dff_descriptions = dff['description'].tolist()
exported_json = []
for i in range(len(dff_ids)):
    exported_json.append({
        "id": dff_ids[i],
        "description": dff_descriptions[i]
    })
with open('hard_problems.json', 'w') as f:
    json.dump(exported_json, f)
```
