```python
import pandas as pd
import json
```

```python
df_all = pd.read_csv('leetcode_dataset.csv')
df_all = df_all[['id', 'description', 'difficulty']]
```

```python
df_all_ids = df_all['id'].tolist()
df_all_descriptions = df_all['description'].tolist()
all_problems_json = []
for i in range(len(df_all_ids)):
    all_problems_json.append({
        "id": df_all_ids[i],
        "description": df_all_descriptions[i]
    })
with open('all_problems.json', 'w') as f:
    json.dump(all_problems_json, f)
```

```python
df_hard = df_all[df_all['difficulty'] == 'Hard']
```

```python
df_hard_ids = df_hard['id'].tolist()
df_hard_descriptions = df_hard['description'].tolist()
hard_problems_json = []
for i in range(len(df_hard_ids)):
    hard_problems_json.append({
        "id": df_hard_ids[i],
        "description": df_hard_descriptions[i]
    })
with open('hard_problems.json', 'w') as f:
    json.dump(hard_problems_json, f)
```
